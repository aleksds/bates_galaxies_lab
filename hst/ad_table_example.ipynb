```python
table = fits.open('J1039_sdsswise.fits')
table[1].data
```

```python
table[1].header
```

```python
# four WISE magnitudes
w1_mag = table[1].data['w1_mag'][0]
w2_mag = table[1].data['w2_mag'][0]
w3_mag = table[1].data['w3_mag'][0]
w4_mag = table[1].data['w4_mag'][0]
print(w1_mag, w2_mag, w3_mag, w4_mag)
```

```python
# magnitude uncertainties
w1_mag_err = table[1].data['w1_mag_err'][0]
w2_mag_err = table[1].data['w2_mag_err'][0]
w3_mag_err = table[1].data['w3_mag_err'][0]
w4_mag_err = table[1].data['w4_mag_err'][0]
print(w1_mag_err, w2_mag_err, w3_mag_err, w4_mag_err)
```

```python
# determining fluxes and uncertainties for W3
w3_nanomaggies = table[1].data['w3_nanomaggies'][0]
w3_nanomaggies_ivar = table[1].data['w3_nanomaggies_ivar'][0]
w3_unc = 1. / np.sqrt(w3_nanomaggies_ivar)
w3_snr = w3_nanomaggies / w3_unc
print(w3_nanomaggies, w3_unc, w3_snr)
```

```python
# check on the conversion between nanomaggies and magnitude
w3_mag_check = -2.5*np.log10(w3_nanomaggies*1e-9)
print(w3_mag_check)
```

```python
# determining fluxes and uncertainties for W4
w4_nanomaggies = table[1].data['w4_nanomaggies'][0]
w4_nanomaggies_ivar = table[1].data['w4_nanomaggies_ivar'][0]
w4_unc = 1. / np.sqrt(w4_nanomaggies_ivar)
w4_snr = w4_nanomaggies / w4_unc
print(w4_nanomaggies, w4_unc, w4_snr)
```

```python
# what if we took the W4 uncertainty as an upper limit on the flux?
w4_upper_1sigma = w4_unc
w4_mag_upper_1sigma = -2.5*np.log10(w4_upper_1sigma*1e-9)
print(w4_mag_upper_1sigma)
```

```python
# what if we treated the negative value as the flux?
w4_mag_perhaps = -2.5*np.log10(np.abs(w4_nanomaggies)*1e-9)
print(w4_mag_perhaps)
```

```python
# difference between upper limit and "value"
w4_err = w4_mag_perhaps - w4_mag_upper_1sigma
print(w4_err)
# so maybe we quote this as 11.73 +/- 1.53 in terms of Vega magnitude?
```

```python
# check values related to chi-squared
w4_prochi2 = table[1].data['w4_prochi2'][0] # Profile-weighted chi-squared
w4_pronpix = table[1].data['w4_pronpix'][0] # number-of-pixels values for chi-squared calculation
print(w4_prochi2, w4_pronpix)
```

```python
# check info about nearby sources
w4_proflux = table[1].data['w4_proflux'][0] #  profile-weighted, the amount of flux contributed by other nearby sources.
w4_profracflux = table[1].data['w4_profracflux'][0] #  equal to w4 proflux divided by w4 nanomaggies -- fraction of flux from other sources
print(w4_proflux, w4_profracflux, w4_proflux/w4_profracflux)
```

```python
# check info on pixels in fit and exposures
w4_npix = table[1].data['w4_npix'][0] # number of pixels included in the fit
w4_pronexp = table[1].data['w4_pronexp'][0] # The number of WISE exposures included in the unWISE coadd at location of source
print(w4_npix, w4_pronexp)
```
