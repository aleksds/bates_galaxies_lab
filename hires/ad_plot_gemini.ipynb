```python
gemgalfol = 'Gemini/'
gemfiles = [f for f in listdir(gemgalfol)]
gemfiles = np.sort(gemfiles)
print(np.sort(gemfiles))
print(len(gemfiles))
print(len(mygalaxies))
mygalaxies['J0826+4305']['redshift']
gemfiles[0][0:10]
```

```python
filename = 'ad_gemini_specta.pdf'

with PdfPages(filename) as pdf:
    
    for i in range(0,len(gemfiles)):
        print(i, gemgalfol+gemfiles[i])
        hdu = fits.open(gemgalfol+gemfiles[i])
        
        maintable = hdu[1].data
        fluxdata = maintable['flux']
        wavelength = maintable['wavelength']
        error = maintable['error']

        name = gemfiles[i][0:10]
        redshift = mygalaxies[name]['redshift']
        print(redshift)
        
        fig = plt.figure()
        ax = fig.add_subplot(1,1,1)
    
        ax.plot(wavelength, fluxdata)
    
        #ax.set_xlim(3800, 9200)
        #ax.set_ylim(0,20)
        ax.set_xlabel(r'Wavelength [$\AA$]', fontsize=13)
        ax.set_ylabel(r'$F_{\lambda}$ [$10^{-17}$ erg s$^{-1}$ cm$^{-2}$ $\AA^{-1}$]', fontsize=13)
        ax.set_title(gemfiles[i])
        
        # mark the halpha line
        plt.axvline(x=6563*(1+redshift), color='red', linestyle='--')
    
        pdf.savefig()
        plt.close()
        
```
