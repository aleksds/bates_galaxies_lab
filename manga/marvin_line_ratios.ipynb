```python
maps = Maps(plateifu='7975-9102')
maps.datamodel
```

```python
ha = maps.emline_gflux_ha_6564
fig = plt.figure(constrained_layout=False, figsize=(15,15))
ax1 = fig.add_subplot(2,2,1)
ha.plot(fig=fig, ax=ax1)
nii = maps.emline_gflux_nii_6585
nii_ha = nii/ha
ax2 = fig.add_subplot(2,2,2)
nii_ha.plot(fig=fig, ax=ax2, title='[N II] / H-alpha')
oi = maps.emline_gflux_oi_6302
oi_ha = oi/ha
ax3 = fig.add_subplot(2,2,3)
oi_ha.plot(fig=fig, ax=ax3, title='[O I] / H-alpha')
hb = maps.emline_gflux_hb_4862
oiii = maps.emline_gflux_oiii_5008
oiii_hb = oiii/hb
ax4 = fig.add_subplot(2,2,4)
fig, ax = oiii_hb.plot(fig=fig,ax=ax4,title='[O III] / H-beta')
plt.tight_layout()
```
